# Data Cleaning


The values in the two columns Status and notes are all null, so we can delete those two columns

```python
df.drop(['Status','Notes'],axis=1,inplace=True)
df
```

Judging by the high number of missing values in some columns like basepay and benefits, it would be impractical to drop entire rows with missing values, a simple strategy to use would be imputing the missing values with the column mean


```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
columns_to_impute = ['Benefits','BasePay','OvertimePay','OtherPay']
imputer.fit(df[columns_to_impute])
df[columns_to_impute] = imputer.transform(df[columns_to_impute])
```

#Basic Data Visualization

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['TotalPay'],bins=20)
plt.show()
```

Plotting proportions of employees per department

```python
# extract department using regular expression by detecting parentheses
df['Department'] = df['JobTitle'].str.extract(r'\((.*?)\)', expand=False)
df['Department'].unique()
```

Some departments have very similar names, we can deal with this redundancy by standardizing similar column names with a single label.

```python
df_cleaned = df.dropna(subset=['Department']) # dropping NANs

df['Department'] = df['Department'].str.upper()

# replace variations with standardized names
df['Department'].replace({
    'CIVIL/CRIMINAL': 'CIVIL & CRIMINAL',
    'CVL&CRMNL': 'CIVIL & CRIMINAL',
    'FIRE DEPT': 'FIRE DEPARTMENT',
    'POLICE DEPARTMENT': 'POLICE DEPARTMENT',
    'MFCC': 'UNKNOWN'
}, inplace=True)
df['Department'].unique()
```

```python
# Plotting
colors = ['skyblue', 'lightgreen', 'lightcoral', 'lightyellow']

plt.figure(figsize=(10, 10))
df_cleaned['Department'].value_counts().plot(
    kind='pie', autopct='%1.1f%%', startangle=90, colors=colors
)
plt.title('Distribution of Job Titles by Department specified')
plt.ylabel('')
plt.show()
```

Just to mention that this method is not a perfect representation of department data as not all entries in JobTitle have the department name explicitly mentioned

#Simple Correlation Analysis

```python
# Plotting the relationship between salary and overtime pay
plt.figure(figsize=(10, 6))
plt.scatter(df['OvertimePay'], df['TotalPay'], color='blue',s=10)
plt.title('Salaries and overtime pay')
plt.xlabel('Overtime pay')
plt.ylabel('Salary')
plt.show()
```

There is a positive correlation between overtime pay and the total salary, the total salary increases with the overtime pay, and if there is no overtime pay the salary remains unchanged, other than a few outliers.

```python
# Plotting the relationship between salary and year
plt.figure(figsize=(10, 6))
plt.scatter(df['TotalPay'], df['Year'], color='blue',s=10)
plt.title('Salaries per year')
plt.xlabel('Salary')
plt.ylabel('Year')
plt.show()
```
